```python
fft_t = np.arange(N)

def plot_signal_fft(signal, axis, sampling_rate, xlim=1000, ylim=16_000):
    N = len(signal)
    T = N / sampling_rate
    
    t = np.arange(len(signal))
    axis[0].plot(t, signal)

    axis[0].set_title('phase1')

    axis[0].set_xlabel('t')
    axis[0].set_ylabel('Vo')

    signal_fft = fft(signal)
    freq = fftfreq(N, 1 / sampling_rate)
    axis[1].stem(freq, np.abs(signal_fft))
    axis[1].set_title('phase1')
    axis[1].set_xlabel('F (Hz)')
    axis[1].set_ylabel('| Vo(F) |')

    axis[1].set_xlim(0, xlim)
    axis[1].set_ylim(0, ylim)
```

```python
from functools import partial

def apply_abs_fft_(sampling_rate, signal):
    fault_type = signal['fault_type']
    signal_cp = signal[:-1].to_numpy() # excluding the label
    
    N = len(signal_cp)
    T = N / sampling_rate
    
    t = np.arange(len(signal_cp))

    signal_fft = fft(signal_cp)
    
    return pd.Series([*abs(signal_fft), fault_type])
    
apply_abs_fft = partial(apply_abs_fft_, sr)

voltages_fft = voltages_stacked_ds.apply(apply_abs_fft, axis=1)
currents_fft = currents_stacked_ds.apply(apply_abs_fft, axis=1)
```

## applying DWT

```python
import pywt
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_scales = 400
scales = range(1,n_scales+1)
waveletname = 'morl'

n_signals = len(voltages_stacked_ds)
voltages_cwt = np.ndarray(shape=(n_signals, n_scales, n_scales))
 
for i, row in enumerate(voltages_stacked_ds.iterrows()):
    signal = row[:-1]
    coeff, freq = pywt.cwt(signal, scales, waveletname, 1)
    coeff_ = coeff[:,:n_scales]
    voltages_cwt[i, :, :] = coeff_

currents_cwt = np.ndarray(shape=(n_signals, n_scales, n_scales))
for i, row in enumerate(currents_stacked_ds.iterrows()):
    signal = row[:-1]
    coeff, freq = pywt.cwt(signal, scales, waveletname, 1)
    coeff_ = coeff[:,:n_scales]
    currents_cwt[i, :, :] = coeff_
```

```python
n_dwt_to_show = 10
fig, ax = plt.subplots(n_dwt_to_show, 2, figsize=(2*10, 5*n_dwt_to_show))

plt.gray()

for i in range(n_dwt_to_show):
    rand_idx = np.random.randint(len(voltages_cwt))
    
    
    voltage_signal_row = voltages_stacked_ds.iloc[rand_idx]
    fault_type = voltage_signal_row['fault_type']
    signal = voltage_signal_row.iloc[:-1]
    
    signal_dwt = voltages_cwt[rand_idx, :, :]
    class_label = class_maps[str(int(fault_type))]
    
    ax[i][0].plot(signal)
    ax[i][0].set_title(f'voltage - {class_label}')
    
    ax[i][1].imshow(signal_dwt)
```
